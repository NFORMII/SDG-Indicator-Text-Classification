```python

experiment_results = []

def evaluate_and_log(exp_name, pipeline, X_test, y_test, y_pred=None):
    """Helper function to predict, calculate metrics, and save results."""
   
    if y_pred is None:
        y_pred = pipeline.predict(X_test)
        
    # Calculate ingMetrics
    h_loss = hamming_loss(y_test, y_pred)
    micro_f1 = f1_score(y_test, y_pred, average='micro')
    macro_f1 = f1_score(y_test, y_pred, average='macro')
    subset_acc = accuracy_score(y_test, y_pred) # Strict exact match accuracy
    
    # Save to the global dictionary
    experiment_results.append({
        'Experiment': exp_name,
        'Hamming Loss': h_loss,
        'Micro F1': micro_f1,
        'Macro F1': macro_f1,
        'Subset Accuracy': subset_acc
    })
    
    #
    print(f"{exp_name} -> HL: {h_loss:.4f} | Micro F1: {micro_f1:.4f} | Macro F1: {macro_f1:.4f} | Subset Acc: {subset_acc:.4f}")

print("Helper function successfully defined. Ready to run experiments.")
```
